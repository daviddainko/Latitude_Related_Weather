Looking at the data, it is clear that latitude and temperature are inversely proportional; as one increases the other decreases.In addition, once the equator is crossed the temperature begins to rise again.

There does not appear to be a correlation between latitude and cloudiness or humidity, as these two can be affected by factors not accounted for with the current data (i.e. local geography, nearby water sources, etc.).

Wind speed does appear to increase at the higher latitudes but, seeing as there are additional factors influenceing this phenomenon as well as insufficient amounts of instances, there is not enough in the current results to call this a trend.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as ow

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
title = ['Name', 'Latitude', 'Temperature', 'Humidity', 'Cloudiness', 'Windspeed']
df_cities = pd.DataFrame(columns=title)
df_cities
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
df_cities.at[:,'Name'] = cities

# Print the city count to confirm sufficient count
len(cities)
```

```python
df_cities
```

```python
units = 'imperial'
settings = {'appid': api_key,'units': units}
city_count = 0

print('Extracting Data:\n---------------------')


for index, row in df_cities.iterrows():
    try:
        response = ow.get_current(row['Name'], **settings)
        lat = ['coord.lat']
        temp = ['main.temp']
        humid = ['main.humidity'] 
        cloud = ['clouds.all']
        wind = ['wind.speed']
        city_count += 1

        print(f"City #%s \nRetrieving data for %s\n" % (str(city_count), str(row['Name'])))
        df_cities.at[index, "Latitude"] = response(*lat)
        df_cities.at[index, "Temperature"] = response(*temp)
        df_cities.at[index, "Humidity"] = response(*humid)
        df_cities.at[index, "Cloudiness"] = response(*cloud)
        df_cities.at[index, "Windspeed"] = response(*wind)

    except:
        print(f"Weather data does not exist for {row['Name']}\n")
        continue
```

```python
df_cities.to_csv('df_cities.csv')
```
